```python
# prompt: add two images

from PIL import Image
image1 = Image.open('/content/trucks.jpg')
image2 = Image.open('/content/truck.jpg')
blended = Image.blend(image1, image2, 0.5)
blended.save('/content/blended.jpg')
```

```python
# prompt: add two images with opencv

import cv2
import numpy as np

image1 = cv2.imread('/content/trucks.jpg')
image2 = cv2.imread('/content/truck.jpg')

added = cv2.add(image1, image2)

cv2.imwrite('/content/added.jpg', added)
```

```python
# prompt: subtraction of two images

image1 = cv2.imread('/content/trucks.jpg')
image2 = cv2.imread('/content/truck.jpg')

subtracted = cv2.subtract(image1, image2)

cv2.imwrite('/content/subtracted.jpg', subtracted)
```

```python
# prompt: multiplication of two images

image1 = cv2.imread('/content/trucks.jpg')
image2 = cv2.imread('/content/truck.jpg')

multiplied = cv2.multiply(image1, image2)

cv2.imwrite('/content/multiplied.jpg', multiplied)
```

```python
from PIL import Image, ImageChops

# creating a image1 object
im1 = Image.open('/content/trucks.jpg')

# creating a image2 object
im2 = Image.open('/content/truck.jpg')

# applying multiply method
im3 = ImageChops.multiply(im1, im2)

im3.show()
im3.save('/content/multiplied.jpg')
```

```python
# prompt: divide image by 2

image1 = cv2.imread('/content/trucks.jpg')
divided = cv2.divide(image1, 0.5)
cv2.imwrite('/content/divided.jpg', divided)
```

```python
from numpy import asarray
from PIL import Image
import matplotlib.pyplot as plt

# load image
image1 = Image.open('/content/trucks.jpg')
pic1 = asarray(image1)
# convert from integers to floats
pic1 = pic1.astype('float32')
# normalize to the range 0-1
pic1 /= 255.0
plt.imshow(pic1)

image2 = Image.open('/content/truck.jpg')
pic2 = asarray(image2)
# convert from integers to floats
pic2 = pic2.astype('float32')
# normalize to the range 0-1
pic2 /= 255.0
plt.imshow(pic2)
#plt.show()

pic3 = (pic1+pic2)*255
plt.imshow(pic3)

pic4 = (pic1*pic2)*255
plt.imshow(pic4)
```
